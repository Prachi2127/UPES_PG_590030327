## What is a Module?
A module is simply a Python file (.py) that contains:

1. functions
2. variables
3. classes
4. reusable code

## Why use modules?

1. Code reuse (write once, use everywhere)
2. Organization (split big programs into smaller parts)
3. Maintainability (easy debugging and updates)
4. Collaboration (different team members can work on different modules) 

## Importing Modules
When you import something, Python:

1. finds the module
2. runs it once
3. makes its names available

```python
import math
print(math.sqrt(16))

from math import sqrt
print(sqrt(16))

import math as m
print(m.pi)
```

## What is a Package?

A package is a folder containing multiple related modules, usually with an __init__.py file.
Suppose you have

basic.py
advanced.py
report.py

Instead of keeping them separately

Put them into one folder

calculator/
    basic.py
    advanced.py
    report.py

This folder becomes a package.

## Why packages?

1. Better structure for large projects
2. Avoid naming conflicts
3. Group related modules together (e.g., finance, utils, db)
