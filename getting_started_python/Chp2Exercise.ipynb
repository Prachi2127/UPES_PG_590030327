## Functions
A function is a block of reusable code that performs a specific task. Instead of writing the same code again and again, we write it once inside a function and call it whenever needed.

__Advantages:__

✅ Code Reusability
✅ Easy to Read
✅ Easy to Debug
✅ Easy to Maintain

## Types of Functions
__1. Built-in Functions__

Already provided by Python.

```python
print()
len()
type()
input()
sorted()
sum()
```

__2. User-defined Functions__
Functions created by the programmer.

```python
def add(a, b):
    return a + b
```

## Variable-Length Positional Arguments (*args)
Suppose you don't know how many numbers the user will pass.

```python
def g(*args):
    return sum(args)
print(g(1,2,3))
```

## Variable-Length Keyword Arguments (**kwargs)


```python
def h(**kwargs):
    return kwargs
print(h(name="Prachi", age=24))
```

__kwargs__ is actually a dictionary.

## Variable Scope (LEGB Rule)
Python searches for variables in this order:

Letter	    Meaning
  L	         Local
  E	       Enclosing
  G	        Global
  B	        Built-in

## Local Variable

Declared inside a function.

```python
x = 10

def foo():
    x = 5
    return x
print(foo())
print(x)
```

## Global Variable

Declared outside all functions.

```python
x = 100

def show():
    print(x)

show()
```
