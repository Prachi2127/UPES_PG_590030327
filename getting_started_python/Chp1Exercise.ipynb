## Dynamic types in Python
Python is a dynamically typed language, which means you do not need to declare the data type of a variable. Python automatically determines the type based on the value assigned to the variable.

```python
x = 10          # Integer
print(type(x))

x = 3.14        # Float
print(type(x))

x = "Hello"     # String
print(type(x))

x = True        # Boolean
print(type(x))
```

Here, the same variable x stores different types of values. Python automatically changes its type.

## Python is an Object Oriented Language
Python is an object-oriented programming (OOP) language because everything in Python is an object. Every object has:

1. Data (attributes/properties)
2. Behavior (methods/functions)

Even simple values like numbers and strings are objects.

Examples of Objects in Python:

Integers → 10
Strings → "hello"
Lists → [1, 2, 3]
Functions → print(), len()
Classes → str, list
Modules → math, os

## Objects in Python
An object is an instance of a class. It is a real entity that contains data (attributes) and functions (methods).
In Python, everything is an object, including numbers, strings, lists, dictionaries, and functions.

Example -->

```python
class Student:
    def __init__(self, name):
        self.name = name

student1 = Student("Prachi")
print(student1.name)
```

Here:

1. Student is a class.
2. student1 is an object (instance) of the Student class.

## How objects are called/accessed in python?
1. In Python, the dot operator (.) is used to access an object's attributes and methods.

```python
text = "hello"
print(text.upper())
```

How python interprets it -->
When we write :

```python
"hello".upper()
```

Python understands it as:

1. "hello" → the object
2. upper → the method (stored in the str class)

Internally, it is equivalent to:

```python
str.upper("hello")
```

Python automatically passes the object ("hello") as the first argument to the method.

2. Built-in Functions vs Methods (Internal Working)
(A) Built-in Function (Global / independent)


```python
len("python")
```

 --> Internal meaning:
Python searches len inside the built-in namespace.

Built-ins live in:

1. __builtins__
2. builtins module

So this is basically:

builtins.len("python")
Built-in functions are not tied to one object.

(B) Method (belongs to an object’s class)

```python
"python".upper()
```

Internal meaning:
Python does:

look at object type → str
find upper inside str
bind it with the object "python"
then call it
Equivalent internal call:

str.upper("python")
So, method = function that becomes bound to an object.
