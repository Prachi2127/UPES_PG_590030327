## <u>Question 8</u> 🔺

> Find area of a triangle when lengths of all three sides are given (**Heron's Formula**).

### 📖 Explanation

**Step 1 — Semi-perimeter:**
$$ s = \frac{a + b + c}{2} $$

**Step 2 — Area:**
$$ \text{Area} = \sqrt{s(s-a)(s-b)(s-c)} $$

| Step | Operation |
|------|-----------|
| 1️⃣ | Compute semi-perimeter `s` |
| 2️⃣ | Plug `s`, `a`, `b`, `c` into the square-root formula |
| 3️⃣ | `math.sqrt()` gives the final area |

➡️ Works for **any** triangle, not just right-angled ones — no need to know the height!


````python
import math

a, b, c = 3, 4, 5
s = (a + b + c) / 2
area = math.sqrt(s * (s - a) * (s - b) * (s - c))
print("Area of triangle:", area)
````

## <u>Question 9</u> ⏱️

> Convert given seconds into hours, minutes, and remaining seconds.

### 📖 Explanation — Floor Division `//` and Modulus `%`

| Operator | Meaning | Used For |
|:--------:|---------|----------|
| `//` | Floor Division — gives **whole quotient**, discards remainder | Extracting hours, then minutes |
| `%` | Modulus — gives the **remainder** | Extracting leftover seconds/minutes |

**Logic:**
- `hours = total_seconds // 3600` → how many full hours fit
- `minutes = (total_seconds % 3600) // 60` → remaining seconds converted to minutes
- `seconds = total_seconds % 60` → whatever seconds are left over

➡️ Example: `7384` seconds = **2 hours, 3 minutes, 4 seconds** ⏳


````python
total_seconds = 7384

hours = total_seconds // 3600
minutes = (total_seconds % 3600) // 60
seconds = total_seconds % 60

print(f"{hours} hour(s), {minutes} minute(s), {seconds} second(s)")
````

| Step | Operation | Effect |

## <u>Question 11</u> ➕

> Find sum of first `n` natural numbers.

### 📖 Explanation

<span style="color:#1565c0;"><b>Formula (Gauss's Sum):</b></span>
$$ \text{Sum} = \frac{n(n+1)}{2} $$

- Instead of looping and adding `1 + 2 + 3 + ... + n`, this **direct formula** computes the sum instantly. ⚡
- Integer division `//` ensures the result stays a whole number (the product `n(n+1)` is always even).

````python
n = 10
sum_n = n * (n + 1) // 2
print("Sum of first", n, "natural numbers is", sum_n)
````

````python
print("a b | a&b a|b a^b")
for a in (0, 1):
    for b in (0, 1):
        print(a, b, "|", a & b, " ", a | b, " ", a ^ b)
````

## <u>Question 13</u> ↔️

> Find left shift and right shift values of a given number.

### 📖 Explanation — Shift Operators

| Operator | Name | Effect | Math Equivalent |
|:--------:|------|--------|------------------|
| `<<` | Left Shift | Shifts bits **left**, filling with 0s | Multiplies by $2^{shift}$ |
| `>>` | Right Shift | Shifts bits **right** | Divides by $2^{shift}$ (floor) |

➡️ Example: `10 << 2` = $10 \times 2^2 = 40$, and `10 >> 2` = $10 \div 2^2 = 2$ (floor division).

In binary: `10` = `1010` → left shift by 2 → `101000` = `40` 🔢


````python
num = 10
shift = 2

print("Number:", num)
print("Left shift by 2:", num << shift)
print("Right shift by 2:", num >> shift)
````

## <u>Question 3</u> ⚖️

> Find the greatest among two numbers. If equal, print `numbers are equal`.

### 📖 Explanation — `if-elif-else` Chain

| Condition | Result |
|-----------|--------|
| `a > b` | `a` is greater |
| `b > a` | `b` is greater |
| *(neither)* → `else` | `a` and `b` must be **equal** |

## <u>Question 5</u> 🧮

> Check whether a quadratic equation has real or imaginary roots and display the roots.

### 📖 Explanation

<span style="color:#1565c0;"><b>Equation:</b></span> $ax^2 + bx + c = 0$

**Step 1 — Discriminant:**
$$ D = b^2 - 4ac $$

**Step 2 — Roots (Quadratic Formula):**
$$ x = \frac{-b \pm \sqrt{D}}{2a} $$

| Discriminant `D` | Nature of Roots |
|:-----------------:|------------------|
| `D ≥ 0` | ✅ **Real** roots |
| `D < 0` | 🌀 **Imaginary** (complex) roots |

➡️ We use the **`cmath`** module (instead of `math`) because it can handle **square roots of negative numbers**, returning complex numbers automatically. This means the same code works for both real and imaginary cases without extra branching! 🎯


````python
import cmath

a, b, c = 1, 2, 5

discriminant = b ** 2 - 4 * a * c
root1 = (-b + cmath.sqrt(discriminant)) / (2 * a)
root2 = (-b - cmath.sqrt(discriminant)) / (2 * a)

if discriminant >= 0:
    print("Roots are real")
else:
    print("Roots are imaginary")

print("Root 1:", root1)
print("Root 2:", root2)
````

## <u>Question 2</u> 💪

> Check whether the given number is an Armstrong number.

### 📖 Explanation
An **Armstrong number** (also called a *narcissistic number*) is a number that equals the **sum of its own digits, each raised to the power of the total number of digits**. ✨

<span style="color:#1565c0;"><b>Formula (for an n-digit number):</b></span>
$$ \text{number} = d_1^n + d_2^n + \dots + d_n^n $$

| Step | Operation |
|------|-----------|
| `num_str = str(num)` | Convert number to string to access individual digits |
| `power = len(num_str)` | Total digit count = the power to raise each digit to |
| Loop over each digit | Add `digit ** power` to a running sum |
| Compare | If sum equals original number → **Armstrong number** ✅ |

➡️ Example: `153` → $1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153$ ✔️


````python
num = 153
num_str = str(num)
power = len(num_str)
sum_of_powers = 0

for digit in num_str:
    sum_of_powers += int(digit) ** power

if sum_of_powers == num:
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
````

## <u>Question 4</u> 🔎

> Check whether a given number is prime or not.

### 📖 Explanation
A **prime number** has exactly two factors: `1` and itself. To check primality efficiently:

| Step | Reasoning |
|------|-----------|
| `if num < 2` | Numbers less than 2 (0, 1, negatives) are **never prime** |
| `range(2, int(num**0.5) + 1)` | ⚡ We only need to check divisors **up to √num** — if no factor exists by then, none will exist beyond it either (huge speed optimization!) |
| `if num % i == 0` | If any number in this range divides `num` evenly, it's **not prime** |
| `break` | Stops the loop **immediately** once a factor is found — no need to keep checking |

➡️ Example: To check `29`, we only test divisors `2` to `5` (since √29 ≈ 5.4) instead of all the way to 28! 🚀


````python
num = 29

if num < 2:
    print(num, "is not prime")
else:
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    print(num, "is prime" if is_prime else "is not prime")
````

| Step | Operation |
|------|-----------|

| `frequency.get(key, 0)` | Returns the **current count** for that letter, or `0` if it hasn't appeared yet — avoids a `KeyError` |
| `frequency[key] = ... + 1` | Increments the count for that letter |
| `sorted(frequency)` | Sorts the dictionary's keys **alphabetically** before printing |

➡️ Example: In `"ABaBCbGc"`, letter `'B'` (or `'b'`) appears **3 times** (`B`, `B`, `b`), so the output includes `3B`. 🔤

## <u>Question 8</u> 🎨

> Take two sets and apply various set operations.

### 📖 Explanation — Complete Set Operations Reference

| Operator | Name | Meaning | Example Result (S1, S2 below) |
|:--------:|------|---------|--------------------------------|
| `\|` | Union | All elements from **either** set | Everything combined, no duplicates |
| `&` | Intersection | Elements common to **both** sets | `{blue}` |
| `S1 - S2` | Difference | In `S1` but **not** in `S2` | Colors unique to S1 |
| `S2 - S1` | Difference (reverse) | In `S2` but **not** in `S1` | Colors unique to S2 |
| `^` | Symmetric Difference | In **either** set, but **not both** (i.e., Union minus Intersection) | Everything **except** the shared elements |

Given:
```
S1 = {Red, yellow, orange, blue}
S2 = {violet, blue, purple}
```

➡️ Since `"blue"` is the only common color, the **symmetric difference** (`^`) will contain everything **except** `"blue"`. 🎨


````python
S1 = {"Red", "yellow", "orange", "blue"}
S2 = {"violet", "blue", "purple"}

print("Union:", S1 | S2)
print("Intersection:", S1 & S2)
print("Difference (S1 - S2):", S1 - S2)
print("Difference (S2 - S1):", S2 - S1)
print("Symmetric Difference:", S1 ^ S2)
````

## <u>Question 1</u> 🔢

> Scan `n` values in range `0-3` and print number of times each value has occurred.

### 📖 Explanation — Dictionary as a Counter

| Step | Operation |
|------|-----------|
| `counts = {0:0, 1:0, 2:0, 3:0}` | Initialize a dictionary with all possible values as **keys**, starting counts at `0` |
| `if value in counts:` | Checks the value is a valid key before updating (safety check) |
| `counts[value] += 1` | Increments the counter for that specific value each time it appears |
| Final `for` loop | Prints the tally for **each** value in the dictionary |

➡️ A dictionary is perfect here because it lets us map each **distinct value → its count** using fast key lookups, instead of writing separate counter variables for each one. 🗂️


````python
values = [0, 1, 2, 3, 2, 1, 0, 3, 3, 2]
counts = {0: 0, 1: 0, 2: 0, 3: 0}

for value in values:
    if value in counts:
        counts[value] += 1

for key in counts:
    print(f"{key} occurred {counts[key]} time(s)")
````

## <u>Question 3</u> 🥈

> Input scores of students and print runner-up score.

### 📖 Explanation

| Step | Operation |
|------|-----------|
| `set(scores)` | Removes **duplicate** scores (e.g., two students scoring `6` count as one distinct value) |
| `sorted(..., reverse=True)` | Sorts the **unique** scores from **highest to lowest** |
| `unique_scores[1]` | Index `0` is the **highest** score (topper); index `1` is the **second-highest distinct** score — the runner-up! 🥈 |

➡️ Example: `[2, 3, 6, 6, 5]` → unique = `{2, 3, 5, 6}` → sorted descending = `[6, 5, 3, 2]` → runner-up = `5` (note: the **duplicate 6** doesn't count as a separate runner-up).


````python
scores = [2, 3, 6, 6, 5]
unique_scores = sorted(set(scores), reverse=True)
runner_up = unique_scores[1]
print("Runner-up score:", runner_up)
````

## <u>Question 1</u> 📈📉

> Write a function to find maximum and minimum from a sequence without using built-in `max()` and `min()`.

### 📖 Explanation

| Step | Operation |
|------|-----------|
| `maximum = minimum = sequence[0]` | Start by assuming the **first element** is both the largest and smallest |
| `for value in sequence[1:]` | Loop through the **remaining** elements (skip index 0, already used) |
| `if value > maximum:` | Update `maximum` whenever a **bigger** value is found |
| `if value < minimum:` | Update `minimum` whenever a **smaller** value is found |
| `return maximum, minimum` | Returns **two values at once** as a tuple |

➡️ In Python, `return a, b` automatically packs the two values into a **tuple**, and writing `mx, mn = find_max_min(nums)` **unpacks** them back into separate variables. 📦


````python
def find_max_min(sequence):
    maximum = sequence[0]
    minimum = sequence[0]

    for value in sequence[1:]:
        if value > maximum:
            maximum = value
        if value < minimum:
            minimum = value

    return maximum, minimum

nums = [15, 3, 99, 42, 7]
mx, mn = find_max_min(nums)
print("Maximum:", mx)
print("Minimum:", mn)
````

> Write a function that takes a positive integer and returns sum of cubes of all positive integers smaller than it.

### 📖 Explanation

<span style="color:#1565c0;"><b>Formula:</b></span>
$$ \text{Sum} = 1^3 + 2^3 + 3^3 + \dots + (n-1)^3 $$

| Step | Operation |
|------|-----------|
| `range(1, n)` | Loops from `1` up to (but **not including**) `n` — so only integers **smaller than n** |
| `total += i ** 3` | Adds the **cube** of each number to the running total |
| `return total` | Sends the final sum back to wherever the function was called |

➡️ Example: For `n = 6`, we cube `1, 2, 3, 4, 5` → $1 + 8 + 27 + 64 + 125 = 225$


````python
def sum_of_cubes_below(n):
    total = 0
    for i in range(1, n):
        total += i ** 3
    return total

n = 6
print(f"Sum of cubes below {n}:", sum_of_cubes_below(n))
````

## <u>Question 3</u> 🔁

> Write a recursive function to print numbers from `1` to `n` (no loops).

### 📖 Explanation — What is Recursion?
**Recursion** is when a function **calls itself** to solve a smaller version of the same problem, until it reaches a **base case** that stops the calls. 🪆

| Part | Role |
|------|------|
| `if n == 0: return` | **Base case** — stops the recursion once we've counted all the way down to 0 |
| `print_1_to_n(n - 1)` | **Recursive call** — the function calls itself with a **smaller** value (`n - 1`) |
| `print(n, end=" ")` | This runs **after** the recursive call returns |

### 🧠 Why does it print in ascending order (1, 2, 3, ... n)?
Because the `print(n)` statement is placed **after** the recursive call. This means the function keeps "diving deeper" (calling itself) **before** printing anything — and only starts printing on the way back **up**, printing the smallest numbers first! 🪜

```
print_1_to_n(3)
 └─ print_1_to_n(2)
     └─ print_1_to_n(1)
         └─ print_1_to_n(0) → returns immediately (base case)
         → prints 1
     → prints 2
 → prints 3
```


````python
def print_1_to_n(n):
    if n == 0:
        return
    print_1_to_n(n - 1)
    print(n, end=" ")

print_1_to_n(10)
print()
````

## <u>Question 4</u> 🐇

> Write a recursive function to print Fibonacci series up to `n` terms.

### 📖 Explanation

<span style="color:#1565c0;"><b>Recursive Definition:</b></span>
$$ fib(n) = \begin{cases} 0 & n = 0 \\ 1 & n = 1 \\ fib(n-1) + fib(n-2) & n > 1 \end{cases} $$

| Part | Role |
|------|------|
| `if n == 0: return 0` | **Base case 1** |
| `if n == 1: return 1` | **Base case 2** |
| `return fib(n-1) + fib(n-2)` | **Recursive case** — each term is the sum of the **two previous terms** |

➡️ The outer `for` loop simply calls `fib(i)` for `i = 0` to `terms - 1`, printing each Fibonacci number in sequence.

<span style="color:#c0392b;"><b>⚠️ Note:</b></span> This recursive approach is elegant but **inefficient** for large `n` (it recalculates the same values many times). For big Fibonacci numbers, an **iterative approach** (like in Experiment 4) or memoization would be faster. ⚡


````python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n - 1) + fib(n - 2)

terms = 8
print("Fibonacci series:")
for i in range(terms):
    print(fib(i), end=" ")
print()
````

> Write functions to explain keyword arguments, default arguments, and variable length arguments.

### 📖 Explanation — Three Ways to Pass Arguments

| Type | Description | Example |
|------|-------------|---------|
| 🏷️ **Keyword Arguments** | Arguments passed using `parameter=value` syntax — order **doesn't matter** since Python matches them by name | `student_info(age=20, name="Ravi")` works the same as `student_info("Ravi", 20)` |
| ⚙️ **Default Arguments** | A parameter is given a **fallback value** in the function definition; if the caller doesn't supply it, the default is used | `greet("Riya")` uses the default message; `greet("Riya", "Good morning")` overrides it |
| 📚 **Variable-Length Arguments (`*args`)** | Lets a function accept **any number** of positional arguments, collected into a **tuple** inside the function | `total_marks(78, 85, 90, 88, 76)` — works no matter how many marks are passed |

➡️ These three features make Python functions extremely **flexible** — the same function can be called in different styles depending on the situation! 🧰


````python
# Keyword argument
def student_info(name, age):
    print(f"Name: {name}, Age: {age}")

student_info(age=20, name="Ravi")

# Default argument
def greet(name, message="Welcome to Python Lab"):
    print(f"{name}: {message}")

greet("Riya")
greet("Riya", "Good morning")

# Variable length argument
def total_marks(*marks):
    return sum(marks)

print("Total marks:", total_marks(78, 85, 90, 88, 76))
````

> Store integers in a file and then find the maximum number, the average, and count numbers greater than 100.

### 📖 Explanation

| Step | Operation |
|------|-----------|
| `f"{number}\n"` | Writes each number as **text**, one per line |

➡️ This question highlights a common beginner mistake: forgetting that **everything read from a text file is a string** by default — numeric operations require explicit conversion (`int()` or `float()`). 🔁

## <u>Question 3</u> 🏙️

> `city.txt` stores city details as `city_name population_in_lakhs area_in_sq_km`. Display all details, cities with population > 10 lakhs, and the sum of all areas.

### 📖 Explanation — Parsing Structured Text

| Step | Operation |
|------|-----------|
| `line.split()` | Splits each line on **whitespace**, producing a list of 3 string fields: `[city, population, area]` |
| `if len(parts) == 3:` | A safety check — skips malformed lines that don't have exactly 3 fields |
| `float(parts[1])`, `float(parts[2])` | Converts the numeric text fields into actual **floats** for calculation |
| `parsed_cities.append((city, population, area))` | Stores each row as a **tuple** inside a list — building a mini structured dataset |

### 🧮 Filtering & Aggregating

| Task | Technique |
|------|-----------|
| Cities with population > 10 lakhs | List comprehension: `[city for city, population, _ in parsed_cities if population > 10]` |
| Sum of all areas | Generator expression inside `sum()`: `sum(area for _, _, area in parsed_cities)` |

➡️ The underscore `_` is a Python convention meaning **"I don't need this value"** — used here to skip fields we're not interested in during unpacking. 🙈


````python
city_records = [
    ("Dehradun", 5.78, 308.20),
    ("Delhi", 190.00, 1484.00),
    ("Mumbai", 124.00, 603.40),
    ("Pune", 74.00, 331.26),
    ("Jaipur", 41.00, 484.60),
]

with open("city.txt", "w", encoding="utf-8") as file:
    for city, population, area in city_records:
        file.write(f"{city} {population} {area}\n")

parsed_cities = []
with open("city.txt", "r", encoding="utf-8") as file:
    for line in file:
        parts = line.split()
        if len(parts) == 3:
            city = parts[0]
            population = float(parts[1])
            area = float(parts[2])
            parsed_cities.append((city, population, area))

print("All city details:")
for city, population, area in parsed_cities:
    print(f"{city}: population={population} lakh, area={area} sq km")

high_population_cities = [city for city, population, _ in parsed_cities if population > 10]
total_area = sum(area for _, _, area in parsed_cities)

print("\nCities with population > 10 lakhs:", high_population_cities)
print("Sum of all city areas:", round(total_area, 2), "sq km")
````

## <u>Question 4</u> ⚠️

> Input `N` test cases of `a b`, perform integer division `a/b`, and handle exceptions for `ZeroDivisionError` and `ValueError`.

### 📖 Explanation — `try-except` Basics

| Concept | Meaning |
|---------|---------|
| `try:` | Code that **might fail** goes here |
| `except (ErrorType1, ErrorType2) as error:` | Catches **specific** exception types; multiple types can be grouped in a tuple |
| `ZeroDivisionError` | Raised when dividing by **zero** |
| `ValueError` | Raised when converting invalid text to a number (e.g., `int("$")`) |

### 🧪 Test Cases Walkthrough

| Input (`a b`) | What Happens |
|:-------------:|--------------|
| `1 0` | `1 // 0` → **ZeroDivisionError** caught → error message printed |
| `2 $` | `int("$")` fails → **ValueError** caught → error message printed |
| `3 1` | `3 // 1 = 3` → normal result printed ✅ |

➡️ The key benefit: without `try-except`, the **very first bad input would crash the entire program**. With it, each test case is handled independently, and the program **keeps running**. 🛡️


````python
raw_input_data = "3\n1 0\n2 $\n3 1"

lines = raw_input_data.strip().splitlines()
n = int(lines[0])

for line in lines[1:n + 1]:
    a_text, b_text = line.split()
    try:
        result = int(a_text) // int(b_text)
        print(result)
    except (ZeroDivisionError, ValueError) as error:
        print("Error Code:", error)
````

## <u>Question 5</u> 🧱

> Create multiple suitable exceptions for a file handling program.

### 📖 Explanation — Custom Exceptions

Python lets you define **your own exception types** by creating a class that inherits from `Exception`. This makes error messages **specific and meaningful** to your program's domain, rather than relying only on generic built-in errors. 🏗️

| Custom Exception | Raised When |
|-------------------|-------------|
| `EmptyFileError` | The file exists but contains **no data** |
| `InvalidRecordError` | A line doesn't match the expected `name,marks` format, or marks aren't numeric |

### 🔍 How `read_student_marks()` Works

| Step | Logic |
|------|-------|
| `try: open(...) except FileNotFoundError` | Catches a missing file and **re-raises** it with a friendlier message using `raise ... from error` |
| `if not lines: raise EmptyFileError(...)` | Triggers if the file has **zero valid lines** |
| `parts = line.split(",")` | Each line should have exactly `name,marks` — **2 parts** |
| `if len(parts) != 2: raise InvalidRecordError(...)` | Catches badly formatted lines (e.g., using `-` instead of `,`) |
| `float(parts[1])` inside another `try-except` | Catches marks that aren't valid numbers (e.g., `"abc"`) |

### 🧪 Test Files Created

| File | Content | Expected Result |
|------|---------|------------------|
| `missing.txt` | *(never created)* | `FileNotFoundError` |
| `students_invalid.txt` | `Aditi-88`, `Rohan,abc` | `InvalidRecordError` (wrong separator / non-numeric marks) |
| `students_valid.txt` | `Aditi,88`, `Rohan,76` | ✅ Successfully parsed dictionary |

➡️ **`raise ... from error`** preserves the **original error's traceback** while presenting a clearer, custom message — great practice for debugging real applications! 🔗


````python
class EmptyFileError(Exception):
    pass


class InvalidRecordError(Exception):
    pass


def read_student_marks(file_name):
    try:
        with open(file_name, "r", encoding="utf-8") as file:
            lines = [line.strip() for line in file if line.strip()]
    except FileNotFoundError as error:
        raise FileNotFoundError(f"{file_name} not found.") from error

    if not lines:
        raise EmptyFileError(f"{file_name} is empty.")

    marks_data = {}
    for line in lines:
        parts = line.split(",")
        if len(parts) != 2:
            raise InvalidRecordError(f"Invalid line format: '{line}'")

        name = parts[0].strip()
        try:
            marks = float(parts[1].strip())
        except ValueError as error:
            raise InvalidRecordError(f"Marks must be numeric in line: '{line}'") from error

        marks_data[name] = marks

    return marks_data


with open("students_valid.txt", "w", encoding="utf-8") as file:
    file.write("Aditi,88\nRohan,76\n")

with open("students_invalid.txt", "w", encoding="utf-8") as file:
    file.write("Aditi-88\nRohan,abc\n")

for candidate in ["missing.txt", "students_invalid.txt", "students_valid.txt"]:
    try:
        data = read_student_marks(candidate)
        print(f"{candidate} -> {data}")
    except (FileNotFoundError, EmptyFileError, InvalidRecordError) as error:
        print(f"{candidate} -> {error}")
````

➡️ Here, we create **3 separate `Student` objects**, each with its own name, SAP ID, and marks dictionary — but all following the **same blueprint**. 🧩

## <u>Question 2</u> 📊

> Add constructor-based initialization and implement `display()`, `marks_percentage()`, and `result()`. Also compute class average.

### 📖 Explanation — Encapsulating Behavior in Methods

| Method | Purpose |
|--------|---------|
| `display()` | Prints the student's basic details |
| `marks_percentage()` | Returns `sum(marks) / number_of_subjects` — the student's **percentage** |
| `result()` | Uses `all(mark > 40 for mark in self.marks.values())` — returns `"Pass"` **only if every single subject** is above 40, otherwise `"Fail"` |
| `class_average(records)` | A **standalone function** (not inside the class) that averages the `marks_percentage()` of **all student objects** together |

### 🧠 Why is `class_average` outside the class?
- It doesn't belong to **one** student — it operates on a **list of student objects**. Keeping it as a separate function (rather than a method) reflects that it's a **class-level (collective) statistic**, not an individual student's behavior. 🧮

➡️ Notice: Meera's physics mark is `39` (below 40), so her `result()` correctly returns **"Fail"**, even though her other subjects are fine — `all()` requires **every** condition to hold. ❗


````python
class StudentRecord:
    def __init__(self, name, sap_id, marks):
        self.name = name
        self.sap_id = sap_id
        self.marks = marks

    def display(self):
        print(f"Name: {self.name} | SAP ID: {self.sap_id} | Marks: {self.marks}")

    def marks_percentage(self):
        return sum(self.marks.values()) / len(self.marks)

    def result(self):
        return "Pass" if all(mark > 40 for mark in self.marks.values()) else "Fail"


def class_average(records):
    return sum(record.marks_percentage() for record in records) / len(records)


input_data = [
    ("Riya", "5001", {"phy": 78, "chem": 67, "maths": 81}),
    ("Prachi", "5002", {"phy": 88, "chem": 92, "maths": 85}),
    ("Meera", "5003", {"phy": 39, "chem": 73, "maths": 64}),
]

records = [StudentRecord(name, sap_id, marks) for name, sap_id, marks in input_data]

for record in records:
    record.display()
    print("Percentage:", f"{record.marks_percentage():.2f}%")
    print("Result:", record.result())
    print("-" * 45)

print("Class average percentage:", f"{class_average(records):.2f}%")
````

## <u>Question 3</u> 🌳

> Create programs to implement different types of inheritance.

### 📖 Explanation — What is Inheritance?
**Inheritance** allows a class (the **child**) to reuse attributes and methods from another class (the **parent**), without rewriting the code. ♻️

### 🌲 Types of Inheritance Demonstrated

| Type | Structure | Example Here |
|------|-----------|--------------|
| 🔹 **Single** | One parent → one child | `Person` → `Teacher` |
| 🔹 **Multilevel** | A chain: grandparent → parent → child | `Animal` → `Mammal` → `Dog` |
| 🔹 **Multiple** | A child inherits from **two or more** parents at once | `Writer` + `Speaker` → `AuthorSpeaker` |

### 🔗 How Each Works

| Class Chain | Result |
|-------------|--------|
| `Teacher(Person)` | `Teacher` automatically gets `introduce()` from `Person`, plus its own `subject()` |
| `Dog(Mammal)` where `Mammal(Animal)` | `Dog` gets **both** `breathe()` (from `Animal`) and `feed_milk()` (from `Mammal`), plus its own `bark()` |
| `AuthorSpeaker(Writer, Speaker)` | Gets **both** `write()` and `speak()` — since Python supports inheriting from multiple parent classes simultaneously |

➡️ This shows how inheritance helps build **increasingly specialized classes** while avoiding duplicate code. 🧬


````python
# Single inheritance
class Person:
    def introduce(self):
        print("I am a person.")


class Teacher(Person):
    def subject(self):
        print("I teach Python.")


# Multilevel inheritance
class Animal:
    def breathe(self):
        print("Breathing...")


class Mammal(Animal):
    def feed_milk(self):
        print("Feeding milk...")


class Dog(Mammal):
    def bark(self):
        print("Woof!")


# Multiple inheritance
class Writer:
    def write(self):
        print("Writing an article.")


class Speaker:
    def speak(self):
        print("Speaking at an event.")


class AuthorSpeaker(Writer, Speaker):
    pass


print("Single Inheritance:")
t = Teacher()
t.introduce()
t.subject()

print("\nMultilevel Inheritance:")
d = Dog()
d.breathe()
d.feed_milk()
d.bark()

print("\nMultiple Inheritance:")
as_role = AuthorSpeaker()
as_role.write()
as_role.speak()
````

## <u>Question 4</u> 🔀

> Create a class to implement method overriding.

### 📖 Explanation — Method Overriding
**Method overriding** happens when a **child class** defines a method with the **same name** as one in its parent class — the child's version **replaces** (overrides) the parent's behavior when called on a child object. 🔁

| Class | `describe()` says | `area()` returns |
|-------|--------------------|-------------------|
| `Shape` (parent) | *"This is a generic shape."* | `0` (default, no specific shape) |
| `Rectangle` (child) | *"This is a rectangle."* (**overridden**) | `length × breadth` (**overridden**) |

➡️ Even though `Rectangle` inherits from `Shape`, calling `rectangle.describe()` and `rectangle.area()` runs the **Rectangle's own versions** — not the parent's — because Python always looks at the **most specific (child) class first**. This is the foundation of **polymorphism**. 🧠


````python
class Shape:
    def describe(self):
        print("This is a generic shape.")

    def area(self):
        return 0


class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def describe(self):
        print("This is a rectangle.")

    def area(self):
        return self.length * self.breadth


shape = Shape()
rectangle = Rectangle(10, 5)

shape.describe()
print("Shape area:", shape.area())

rectangle.describe()
print("Rectangle area:", rectangle.area())
````

## <u>Question 5</u> ➕

> Create a class for operator overloading which adds two `Point` objects.

### 📖 Explanation — Operator Overloading with `__add__`

Python lets us **redefine what an operator does** for our own custom classes using special "dunder" (double-underscore) methods. ✨

| Special Method | Purpose |
|----------------|---------|
| `__init__` | Constructor — sets up `x` and `y` for each point |
| `__add__(self, other)` | Defines what happens when you write `p1 + p2` — here, it returns a **new `Point`** whose `x` and `y` are the **sums** of the two points' coordinates |
| `__repr__` | Defines how the object is **displayed** when printed — without it, printing a `Point` would show something unreadable like `<__main__.Point object at 0x...>` |

### 🧮 Walkthrough
```
P1 = Point(10, 20)
P2 = Point(12, 15)
P3 = P1 + P2   →   calls P1.__add__(P2)
              →   Point(10+12, 20+15)
              →   Point(x=22, y=35)
```

➡️ The check `if not isinstance(other, Point): return NotImplemented` is a **safety net** — it ensures we only try to add two `Point` objects together, not a `Point` and some unrelated type. 🛡️


````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if not isinstance(other, Point):
            return NotImplemented
        return Point(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"Point(x={self.x}, y={self.y})"


p1 = Point(10, 20)
p2 = Point(12, 15)
p3 = p1 + p2

print("P1:", p1)
print("P2:", p2)
print("P3 = P1 + P2:", p3)
````

````python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 140, 135, 160, 175, 190]
expenses = [95, 105, 102, 110, 120, 130]

categories = ["Rent", "Food", "Transport", "Others"]
share = [35, 25, 20, 20]

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0, 0].plot(months, sales, marker="o", color="teal")
axes[0, 0].set_title("Line Chart: Monthly Sales")
axes[0, 0].set_ylabel("Sales")
axes[0, 0].grid(True, linestyle="--", alpha=0.5)

axes[0, 1].bar(months, expenses, color="orange")
axes[0, 1].set_title("Bar Chart: Monthly Expenses")
axes[0, 1].set_ylabel("Expenses")

axes[1, 0].scatter(sales, expenses, color="crimson")
axes[1, 0].set_title("Scatter Plot: Sales vs Expenses")
axes[1, 0].set_xlabel("Sales")
axes[1, 0].set_ylabel("Expenses")

axes[1, 1].pie(share, labels=categories, autopct="%1.1f%%", startangle=90)
axes[1, 1].set_title("Pie Chart: Expense Distribution")

plt.tight_layout()
plt.show()
````
